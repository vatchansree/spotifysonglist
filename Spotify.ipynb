```python
df1[['Danceability','Energy','Loudness','Speechiness','Acousticness','Liveness','Tempo','Valence']] = df1[['Danceability','Energy','Loudness','Speechiness','Acousticness','Liveness','Tempo','Valence']].astype('float16')
df1['Popularity'] = df1['Popularity'].astype('int16')
df1['Duration (ms)'] = df1['Duration (ms)'].astype('int16')
df1.info()
```
